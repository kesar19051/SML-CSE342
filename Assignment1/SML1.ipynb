```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
from math import *
```

# Question 1

```python
mu1 = 2
variance1 = 1
sigma1 = math.sqrt(variance1)
x1 = np.linspace(mu1 - 4*sigma1, mu1 + 4*sigma1, 100)
plt.xlabel("x")
plt.ylabel("P(x|w1)")
plt.plot(x1, stats.norm.pdf(x1, mu1, sigma1), color = "green")
plt.show()
```

```python
mu2 = 5
variance2 = 1
sigma2 = math.sqrt(variance2)
x2 = np.linspace(mu2 - 4*sigma2, mu2 + 4*sigma2, 100)
plt.xlabel("x")
plt.ylabel("P(x|w2)")
plt.plot(x2, stats.norm.pdf(x2, mu2, sigma2), color = "red")
plt.show()
```

```python
plt.plot(x1, stats.norm.pdf(x1, mu1, sigma1), color = "green")
plt.plot(x2, stats.norm.pdf(x2, mu2, sigma2), color = "red")
plt.ylabel("P(x|wi)")
plt.xlabel("x")
plt.legend(["P(x|w1)", "P(x|w2)"])
plt.show()
```
